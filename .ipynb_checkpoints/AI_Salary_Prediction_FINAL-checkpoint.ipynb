## 3. 🧹 Data Cleaning — the `wrangle()` function

We wrap all cleaning steps into a single reusable function. This is professional practice: the same cleaning logic must apply to both training data and any new data fed to the deployed model. If cleaning lives in a function, you cannot accidentally apply it inconsistently.

**Decisions made inside `wrangle()`:**

- **Drop `job_id`, `salary_currency`, `company_name`, `posting_date`, `application_deadline`** — see Data Dictionary for justification of each
- **Remove salary outliers using IQR method** — we compute the interquartile range and remove rows where salary falls below Q1–1.5×IQR or above Q3+1.5×IQR. This removes extreme values that would distort the model's learning for the typical salary range. Note: these are not data errors — they are real salaries — but removing them improves prediction accuracy for the majority of cases


```python
def wrangle(df):
    """
    Clean and prepare the AI jobs dataset.
    Apply this function to any data before modelling or prediction.
    """
    df = df.copy()

    # Drop columns with no predictive value or that need special treatment
    df.drop(columns=['job_id', 'salary_currency', 'company_name',
                     'posting_date', 'application_deadline'], inplace=True)

    # Remove salary outliers using IQR method
    Q1 = df['salary_usd'].quantile(0.25)   # 25th percentile
    Q3 = df['salary_usd'].quantile(0.75)   # 75th percentile
    IQR = Q3 - Q1                           # Interquartile range

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[(df['salary_usd'] >= lower) & (df['salary_usd'] <= upper)]

    return df.reset_index(drop=True)


df = wrangle(df_raw)

print(f"Original rows:  {len(df_raw):,}")
print(f"After cleaning: {len(df):,}")
print(f"Rows removed:   {len(df_raw) - len(df):,} (outliers)")
print(f"\nSalary range after cleaning:")
print(f"  Min: ${df['salary_usd'].min():,.0f}")
print(f"  Max: ${df['salary_usd'].max():,.0f}")
print(f"  Mean: ${df['salary_usd'].mean():,.0f}")
```

## 4. 🎯 Baseline Model — Setting the Bar

Before doing any EDA or building a real model, we establish a **baseline**: the simplest possible prediction. Here we predict the average salary for every job, regardless of any features.

**Why this matters:** A real ML model is only valuable if it beats this baseline. If it doesn't, we wasted our time. The baseline gives us a minimum performance bar and lets us calculate the improvement our model delivers.


```python
y = df['salary_usd']

# Baseline: predict the mean salary for every single record
y_mean = y.mean()
y_pred_baseline = [y_mean] * len(y)

baseline_mae = mean_absolute_error(y, y_pred_baseline)

print("=" * 50)
print("  BASELINE MODEL (Predict Mean for Everyone)")
print("=" * 50)
print(f"  Mean salary predicted: ${y_mean:,.0f}")
print(f"  Baseline MAE:          ${baseline_mae:,.0f}")
print()
print("  Interpretation: If we just guess the average salary")
print(f"  for every job, we are wrong by ${baseline_mae:,.0f} on average.")
print()
print("  ✅ Any real model MUST beat this to be useful.")
```

## 6. 🔧 Feature Engineering

### Why we need to engineer features

Machine learning models are mathematical — they only understand numbers. All categorical text columns must be converted to numbers. This section explains every engineering decision made.

### Engineered feature: `same_country`
We create a new binary feature from two existing columns: `company_location` and `employee_residence`. If both are the same country, `same_country = 1`; otherwise `= 0`. This captures whether the employee is a local hire or international — which can influence salary structures.

### Two approaches to categorical encoding

In Section 7 (Manual Encoding), we manually encode all categorical features using ordinal encoding and frequency-rank encoding, then train 5 models. In Section 8 (Pipeline), we let `sklearn`'s `OneHotEncoder` handle all categoricals automatically inside a `make_pipeline`. We compare both to understand the difference.


```python
# ── Engineer the same_country feature ────────────────────────────────────────
df['same_country'] = (df['company_location'] == df['employee_residence']).astype(int)

same_pct = df['same_country'].mean() * 100
print(f"Employees in same country as company: {same_pct:.1f}%")
print(f"Employees in different country:       {100-same_pct:.1f}%")

# Salary difference
same_sal   = df[df['same_country']==1]['salary_usd'].mean()
diff_sal   = df[df['same_country']==0]['salary_usd'].mean()
print(f"\nAvg salary — same country: ${same_sal:,.0f}")
print(f"Avg salary — diff country: ${diff_sal:,.0f}")
print(f"Difference: ${abs(same_sal-diff_sal):,.0f} — justifies including this feature")
```

```python
# ── Manual Ordinal + Frequency Encoding ──────────────────────────────────────
# Used in Round 1 (manual models) to understand encoding concepts explicitly

exp_order  = ['EN','MI','SE','EX']      # Entry < Mid < Senior < Executive
edu_order  = ['Associate','Bachelor','Master','PhD']
size_order = ['S','M','L']             # Small < Medium < Large
emp_order  = ['PT','FL','CT','FT']     # Part-time < Freelance < Contract < Full-time

df['exp_encoded']  = df['experience_level'].map({v:i for i,v in enumerate(exp_order)})
df['edu_encoded']  = df['education_required'].map({v:i for i,v in enumerate(edu_order)})
df['size_encoded'] = df['company_size'].map({v:i for i,v in enumerate(size_order)})
df['emp_encoded']  = df['employment_type'].map({v:i for i,v in enumerate(emp_order)})

# Frequency-rank encoding for high-cardinality columns
top_industries = df['industry'].value_counts().index.tolist()
top_titles     = df['job_title'].value_counts().head(20).index.tolist()

df['industry_encoded'] = df['industry'].apply(
    lambda x: top_industries.index(x) if x in top_industries else -1)
df['title_encoded'] = df['job_title'].apply(
    lambda x: top_titles.index(x) if x in top_titles else 20)

print("Manual encoding complete.")
print()
print("Why ORDINAL encoding for experience_level, education, company_size, employment_type?")
print("  → These have a natural ORDER. Entry < Mid < Senior < Executive is real and meaningful.")
print("  → The numbers 0,1,2,3 genuinely represent 'more' of something.")
print()
print("Why FREQUENCY-RANK encoding for industry and job_title?")
print("  → These have no natural order (Consulting is not 'more' than Technology).")
print("  → One-hot encoding would create 15+ new columns for industry alone — feature explosion.")
print("  → Frequency rank gives a numeric signal without blowing up dimensionality.")
print()
print("Why BINARY feature for same_country?")
print("  → Derived from two columns (company_location & employee_residence) into one meaningful signal.")
```

## 7. 🤖 Round 1 — Manual Encoding Models

We train 5 models using our hand-crafted encoding. This round is important for **learning** — every encoding decision is explicit and visible, so we understand exactly what the model receives.

**Why these 5 models?**
- **Linear Regression** — always start with the simplest baseline model for regression
- **Ridge Regression** — Linear Regression with L2 regularization penalty. Included to test whether plain LR is overfitting. If Ridge improves it significantly, overfitting exists. Similar results = no significant overfitting
- **Random Forest** — ensemble of many decision trees. Handles non-linear relationships and feature interactions without scaling
- **Gradient Boosting** — builds trees sequentially, each correcting the errors of the previous. Generally the strongest performer on tabular data
- **XGBoost** — optimised implementation of gradient boosting. Industry standard for structured data problems


```python
# ── Features and Target ──────────────────────────────────────────────────────
features_manual = ['exp_encoded','edu_encoded','size_encoded','emp_encoded',
                   'remote_ratio','years_experience','job_description_length',
                   'benefits_score','industry_encoded','title_encoded','same_country']

X_manual = df[features_manual]
y        = df['salary_usd']
sal_range = y.max() - y.min()

print(f"Feature matrix shape: {X_manual.shape}")
print(f"Target range: ${y.min():,.0f} — ${y.max():,.0f}")
print(f"Range: ${sal_range:,.0f}")
```

```python
# ── Train/Test Split ─────────────────────────────────────────────────────────
# 80% for training, 20% for testing. random_state=42 ensures reproducibility.
# Stratifying is not needed for regression (only for classification).

X_tr_m, X_te_m, y_tr, y_te = train_test_split(
    X_manual, y, test_size=0.2, random_state=42)

print(f"Training set: {X_tr_m.shape[0]:,} rows")
print(f"Test set:     {X_te_m.shape[0]:,} rows")

# Scaling — required for linear models only
# Linear models assign coefficients to each feature. Without scaling, a feature
# with large values (e.g. job_description_length up to 2500) gets a smaller
# coefficient than one with small values — not because it matters less, but
# because of scale. StandardScaler fixes this by normalising all features to
# mean=0, std=1. Tree-based models split on thresholds — scale does not affect them.
scaler = StandardScaler()
X_tr_m_sc = scaler.fit_transform(X_tr_m)
X_te_m_sc = scaler.transform(X_te_m)

print("\n✅ Split complete. Scaling applied for linear models.")
```

```python
# ── Train all 5 manual models & collect metrics ───────────────────────────────
def evaluate_model(name, approach, y_tr, y_te, pred_tr, pred_te, train_time):
    rows = []
    for split, yt, yp in [('Train', y_tr, pred_tr), ('Test', y_te, pred_te)]:
        mae  = mean_absolute_error(yt, yp)
        mse  = mean_squared_error(yt, yp)
        rmse = np.sqrt(mse)
        r2   = r2_score(yt, yp)
        rows.append({
            'Regressor': name, 'Approach': approach, 'Split': split,
            'MAE': round(mae, 0), 'MSE': round(mse, 0),
            'RMSE': round(rmse, 0), 'R²': round(r2, 4),
            'RMSE % of Range': round(rmse / sal_range * 100, 2),
            'Train Time (s)': round(train_time, 3)
        })
    return rows

manual_results = []

# Linear Regression
t0 = time.time()
lr = LinearRegression()
lr.fit(X_tr_m_sc, y_tr)
t = time.time() - t0
manual_results += evaluate_model('Linear Regression','Manual Encoding',
    y_tr, y_te, lr.predict(X_tr_m_sc), lr.predict(X_te_m_sc), t)

# Ridge Regression
t0 = time.time()
ridge = Ridge(alpha=10)
ridge.fit(X_tr_m_sc, y_tr)
t = time.time() - t0
manual_results += evaluate_model('Ridge Regression','Manual Encoding',
    y_tr, y_te, ridge.predict(X_tr_m_sc), ridge.predict(X_te_m_sc), t)

# Random Forest
t0 = time.time()
rf = RandomForestRegressor(n_estimators=100, max_depth=12, random_state=42, n_jobs=-1)
rf.fit(X_tr_m, y_tr)
t = time.time() - t0
manual_results += evaluate_model('Random Forest','Manual Encoding',
    y_tr, y_te, rf.predict(X_tr_m), rf.predict(X_te_m), t)

# Gradient Boosting
t0 = time.time()
gbm = GradientBoostingRegressor(n_estimators=100, learning_rate=0.05, max_depth=5, random_state=42)
gbm.fit(X_tr_m, y_tr)
t = time.time() - t0
manual_results += evaluate_model('Gradient Boosting','Manual Encoding',
    y_tr, y_te, gbm.predict(X_tr_m), gbm.predict(X_te_m), t)

# XGBoost
t0 = time.time()
xgb_m = XGBRegressor(n_estimators=150, learning_rate=0.05, max_depth=6, random_state=42, verbosity=0)
xgb_m.fit(X_tr_m, y_tr)
t = time.time() - t0
manual_results += evaluate_model('XGBoost','Manual Encoding',
    y_tr, y_te, xgb_m.predict(X_tr_m), xgb_m.predict(X_te_m), t)

manual_df = pd.DataFrame(manual_results)
test_manual = manual_df[manual_df['Split']=='Test'].sort_values('R²', ascending=False)
print("\n=== ROUND 1: Manual Encoding — Test Results ===")
print(test_manual[['Regressor','MAE','MSE','RMSE','R²','RMSE % of Range','Train Time (s)']].to_string(index=False))
```

```python
# ── Visualise Round 1 results ─────────────────────────────────────────────────
test_m = manual_df[manual_df['Split']=='Test'].sort_values('R²')

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
colors = ['#4f8bf9','#34c98c','#f9a825','#e53935','#ab47bc']

for ax, metric in zip(axes, ['R²','MAE','RMSE']):
    vals = test_m[metric].values
    bars = ax.barh(test_m['Regressor'], vals, color=colors, alpha=0.85)
    ax.set_title(f'{metric} — Round 1 (Manual)', fontweight='bold')
    ax.set_xlabel(metric)
    best_idx = vals.argmax() if metric == 'R²' else vals.argmin()
    bars[best_idx].set_edgecolor('gold')
    bars[best_idx].set_linewidth(3)
    for bar, v in zip(bars, vals):
        label = f'{v:.4f}' if metric=='R²' else f'${v:,.0f}'
        ax.text(bar.get_width()*0.02, bar.get_y()+bar.get_height()/2,
                label, va='center', fontsize=9, color='white', fontweight='bold')

plt.suptitle('Round 1 Model Comparison — Manual Encoding (Best model highlighted gold)',
             fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# ── Overfitting check — Round 1 ──────────────────────────────────────────────
print("Overfitting check: comparing Train R² vs Test R²")
print("If Train R² is much higher than Test R², the model memorised training data.")
print()
print(f"{'Model':<25} {'Train R²':>10} {'Test R²':>10} {'Gap':>8} {'Status':>12}")
print("-" * 70)
for model in manual_df['Regressor'].unique():
    tr = manual_df[(manual_df['Regressor']==model)&(manual_df['Split']=='Train')]['R²'].values[0]
    te = manual_df[(manual_df['Regressor']==model)&(manual_df['Split']=='Test')]['R²'].values[0]
    gap = tr - te
    status = "⚠️  OVERFIT" if gap > 0.05 else "✅ OK"
    print(f"{model:<25} {tr:>10.4f} {te:>10.4f} {gap:>8.4f} {status:>12}")
```

## 8. 🚀 Round 2 — make_pipeline with OneHotEncoder

### What is make_pipeline?

`make_pipeline` is not a new model. It is a **conveyor belt** that chains preprocessing steps and a model into a single object. When you call `pipeline.fit(X, y)`, it:

1. Fits the encoder on X
2. Transforms X using those encodings
3. Fits the model on the transformed data

When you call `pipeline.predict(X_new)`:
1. Transforms X_new using the *same* encoder fitted on training data
2. Passes transformed data to the model
3. Returns predictions

This is production-ready: one `.pkl` file, one `.predict()` call. No risk of applying different encoding to new data.

### Why does OHE in a pipeline outperform manual encoding so significantly?

In Round 1, `job_title` was encoded as a single number (0–20). XGBoost then has to learn salary differences between all 20 titles from one column. With OneHotEncoder, each job title gets its own binary column — XGBoost can learn *independently* that "AI Specialist" pays differently from "Data Analyst" without any ordinal assumption between them. The richer representation extracts far more signal.


```python
# ── Define features for pipeline approach ────────────────────────────────────
# Categorical columns → OneHotEncoder creates a binary column for each unique value
# Numeric columns → passed through unchanged (passthrough)

cat_cols = ['job_title', 'experience_level', 'employment_type', 'company_location',
            'company_size', 'education_required', 'industry']
num_cols = ['remote_ratio', 'years_experience', 'job_description_length',
            'benefits_score', 'same_country']

X_pipe = df[cat_cols + num_cols]

# Function to create a fresh preprocessor each time
# (needed because sklearn transformers cannot be reused after fitting)
def make_preprocessor():
    return ColumnTransformer(transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols),
        ('num', 'passthrough', num_cols)
    ])

X_tr_p, X_te_p, _, _ = train_test_split(X_pipe, y, test_size=0.2, random_state=42)

print(f"Pipeline training set: {X_tr_p.shape[0]:,} rows, {len(cat_cols)+len(num_cols)} raw features")

# Show how many features OHE creates
import numpy as np
test_enc = make_preprocessor().fit(X_tr_p)
n_features_after = test_enc.transform(X_tr_p).shape[1]
print(f"After OneHotEncoding: {n_features_after} features (vs 11 in manual approach)")
print(f"\nThis richer feature space is why pipeline models outperform manual encoding models.")
```

```python
# ── Train all 5 pipeline models ───────────────────────────────────────────────
pipe_results = []

models_to_run = [
    ('Linear Regression', LinearRegression()),
    ('Ridge Regression',  Ridge(alpha=10)),
    ('Random Forest',     RandomForestRegressor(n_estimators=100, max_depth=12, random_state=42, n_jobs=-1)),
    ('Gradient Boosting', GradientBoostingRegressor(n_estimators=100, learning_rate=0.05, max_depth=5, random_state=42)),
    ('XGBoost',           XGBRegressor(n_estimators=150, learning_rate=0.05, max_depth=6, random_state=42, verbosity=0)),
]

for name, estimator in models_to_run:
    t0 = time.time()
    pipe = make_pipeline(make_preprocessor(), estimator)
    pipe.fit(X_tr_p, y_tr)
    t = time.time() - t0
    pipe_results += evaluate_model(name, 'Pipeline + OHE',
        y_tr, y_te, pipe.predict(X_tr_p), pipe.predict(X_te_p), t)
    print(f"  ✅ Pipeline {name} trained in {t:.2f}s")

pipe_df = pd.DataFrame(pipe_results)
test_pipe = pipe_df[pipe_df['Split']=='Test'].sort_values('R²', ascending=False)
print("\n=== ROUND 2: Pipeline + OHE — Test Results ===")
print(test_pipe[['Regressor','MAE','MSE','RMSE','R²','RMSE % of Range','Train Time (s)']].to_string(index=False))
```

## 11. 🔄 Cross-Validation — Confirming Our Result Is Reliable

```python
# Cross-validation answers: "Is our R² of 0.884 real, or did we just get lucky
# with this particular train-test split?"
#
# 5-fold CV splits the full dataset into 5 equal parts.
# It trains on 4 parts and tests on 1 part, rotating 5 times.
# Each fold gives a separate R² score. We take the mean and std.
#
# If all 5 scores are close to each other (low std), the result is stable.
# If scores vary wildly, our single split evaluation was unreliable.

print("Running 5-fold cross-validation on XGBoost Pipeline...")
print("(This trains the model 5 separate times — may take 30-60 seconds)")
print()

cv_model = make_pipeline(
    make_preprocessor(),
    XGBRegressor(n_estimators=150, learning_rate=0.05, max_depth=6,
                 random_state=42, verbosity=0)
)

cv_scores = cross_val_score(cv_model, X_pipe, y, cv=5, scoring='r2', n_jobs=-1)

print("=" * 50)
print("  CROSS-VALIDATION RESULTS (XGBoost Pipeline)")
print("=" * 50)
for i, score in enumerate(cv_scores, 1):
    bar = '█' * int(score * 30)
    print(f"  Fold {i}: {score:.4f}  {bar}")
print(f"  {'─'*40}")
print(f"  Mean R²:  {cv_scores.mean():.4f}")
print(f"  Std:      {cv_scores.std():.4f}")
print()
print("✅ Interpretation: All 5 folds give R² between", 
      f"{cv_scores.min():.3f} and {cv_scores.max():.3f}.")
print(f"   Std of {cv_scores.std():.4f} is very small — our result is stable and not lucky.")
print("   The model will generalise reliably to new, unseen job postings.")
```

## 12. ⚙️ Hyperparameter Tuning — GridSearchCV

```python
# GridSearchCV systematically tries every combination of hyperparameters
# you specify and finds which combination gives the best cross-validated score.
#
# We only tune the WINNER (XGBoost Pipeline) — no point tuning losers.
#
# Hyperparameters being tuned:
# - n_estimators: how many trees to build (more = more learning, longer training)
# - learning_rate: how much each tree contributes (lower = more careful, needs more trees)
# - max_depth: how deep each tree grows (deeper = more complex patterns, risk of overfit)

print("Running GridSearchCV...")
print("Testing", 2*2*2, "parameter combinations × 3 CV folds =", 2*2*2*3, "fits total")
print()

param_grid = {
    'xgbregressor__n_estimators':  [100, 200],
    'xgbregressor__learning_rate': [0.05, 0.1],
    'xgbregressor__max_depth':     [4, 6],
}

gs_pipeline = make_pipeline(
    make_preprocessor(),
    XGBRegressor(random_state=42, verbosity=0)
)

gs = GridSearchCV(
    gs_pipeline, param_grid,
    cv=3, scoring='r2', n_jobs=-1, verbose=1
)
gs.fit(X_tr_p, y_tr)

print()
print("=" * 55)
print("  GRIDSEARCHCV RESULTS")
print("=" * 55)
print(f"  Best parameters found:")
for param, val in gs.best_params_.items():
    print(f"    {param.split('__')[1]}: {val}")
print(f"  Best cross-validated R²: {gs.best_score_:.4f}")

# Evaluate tuned model on test set
tuned_pred = gs.predict(X_te_p)
tuned_mae  = mean_absolute_error(y_te, tuned_pred)
tuned_rmse = np.sqrt(mean_squared_error(y_te, tuned_pred))
tuned_r2   = r2_score(y_te, tuned_pred)

print(f"\n  Tuned Model — Test Set Performance:")
print(f"    MAE:            ${tuned_mae:,.0f}")
print(f"    RMSE:           ${tuned_rmse:,.0f}")
print(f"    R²:             {tuned_r2:.4f}")
print(f"    RMSE % of Range:{tuned_rmse/sal_range*100:.2f}%")
print(f"\n  Improvement over pre-tuned XGBoost:")
print(f"    MAE change:     ${12912-tuned_mae:+,.0f}")
print(f"    R² change:      {tuned_r2-0.8844:+.4f}")
```

6. **Pipeline + OHE dramatically outperforms manual encoding** — R² improved from 0.636 to 0.884 by letting OneHotEncoder preserve the full information in categorical features.
7. **Our model is not overfitting** — The small gap between Train R² (0.910) and Test R² (0.884) confirms the model generalises well to unseen data.
8. **Cross-validation confirms reliability** — 5-fold CV gives consistent scores (std=0.004), proving our result is not a lucky split.
9. **68%+ improvement over baseline** — The model reduces prediction error from ~$41K (naive mean) to ~$12.9K — a meaningful, deployable improvement.

### What the remaining error tells us
An R² of 0.884 means 11.6% of salary variance is still unexplained. This is expected — salary is influenced by factors not in job postings: individual negotiation ability, internal pay bands, cost-of-living adjustments, equity compensation, and regional economic conditions. No job posting dataset will explain 100% of salary.
